**SUICIDE IDEATION**

1. Import Incident Data
2. Merge with All Dates -- Join on Data
3. Add Final Day to ENTER - 1 Day, 10 Days from All Dates table
4. Create field: Entered on Time: If OccDateEntered <= Final Day to Enter - 1 Day --> YES, otherwise, NO
5. Create field: Updated on Time: If Updated Date <= 10 Days --> YES, otherwise, NO
6. If Updated Data is Blank, leave Updated on Time Blank.

```python
#schools_susptype = schools.merge(susptypefinal, left_on='DBN', right_on='ATS', how='left')
ideation_df = ideation.merge(alldates, left_on='PrincInformed', right_on='Date', how='left')
```

```python
#schools_susptype.drop('ATS', axis = 1, inplace = True)
#Drop Day of Week, School Day, Nearest School Day, 2 Days, 5 Days
ideation_df.drop('Day of Week', axis = 1, inplace = True)
ideation_df.drop('School Day', axis = 1, inplace = True)
ideation_df.drop('Nearest School Day', axis = 1, inplace = True)
ideation_df.drop('2 Days', axis = 1, inplace = True)
ideation_df.drop('5 Days', axis = 1, inplace = True)
ideation_df.drop('UpdatedText', axis = 1, inplace = True)
ideation_df.drop('EventType', axis = 1, inplace = True)
ideation_df.drop('Date_x', axis = 1, inplace = True)
ideation_df.drop('Date_y', axis = 1, inplace = True)
ideation_df.drop('RowNumber', axis = 1, inplace = True)
```

```python
ideation_df
```

```python
#Create fileds: Entered on Time: If OccDateEntered <= Final Day to Enter - 1 Day --> YES, otherwise, NO
#Source: https://www.geeksforgeeks.org/python-creating-a-pandas-dataframe-column-based-on-a-given-condition/
#df['hasimage'] = np.where(df['photos']!= '[]', True, False)  
#Source: https://www.dataquest.io/blog/tutorial-add-column-pandas-dataframe-based-on-if-else-condition/
#df['Price'] = [1500 if x =='Music' else 800 for x in df['Event']] 

ideation_df['Entered on Time'] = np.where(ideation_df['OccDateEntered'] <= ideation_df['Final Day to ENTER - 1 Day'], 'YES', 'NO')
```

```python
ideation_df
```

```python
ideation_df['Updated on Time'] = np.where(ideation_df['Updated Date'] <= ideation_df['10 Days'], 'YES', 'NO')
```

```python
ideation_df
```

```python
#Updated on Time: If Updated Date <= 10 Days --> YES, otherwise, NO
#ideation_df['Entered on Time'] = np.where(ideation_df['OccDateEntered'] <= ideation_df['Final Day to ENTER - 1 Day'], 'YES', 'NO')
ideation_df['Updated on Time'] = np.where(ideation_df['Updated Date'].isnull(), ' ', ideation_df['Updated on Time'])
```

**MATERIAL INCIDENTS**

```python
material_df = material.merge(alldates, left_on='PrincInformed', right_on='Date', how='left')
```

```python
material_df.info()
```

```python
material_df['Entered on Time'] = np.where(material_df['OccDateEntered'] <= material_df['Final Day to ENTER - 1 Day'], 'YES', 'NO')
```

```python
material_df['Investigated on Time'] = np.where(material_df['CRA832InitialInvestigationCompletionDate'] <= material_df['5 Days'], 'YES', 'NO')
```

```python
material_df['Investigated on Time'] = np.where(material_df['CRA832InitialInvestigationCompletionDate'].isnull(), ' ', material_df['Investigated on Time'])
```

```python
material_df['Updated on Time'] = np.where(material_df['Update Date'] <= material_df['10 Days'], 'YES', 'NO')
```
