1. Import Incident Data
2. Merge with All Dates -- Join on Data
    Add Final Day to ENTER - 1 Day, 10 Days 
3. Create fileds: Entered on Time: If OccDateEntered <= Final Day to Enter - 1 Day --> YES, otherwise, NO
                   Updated on Time: If Updated Date <= 10 Days --> YES, otherwise, NO

```python
import pandas as pd
import numpy as np
```

```python
###Import data
##1. Find file in finder
##2. Option + Command + C
##3. Paste Info read_csv(r” ”)

df = pd.ExcelFile(r"/Users/holly/Desktop/nycdoe/01142021 Suicide Data.xlsx")
```

```python
#Read data from corrrect tab.
#schools = suspmaster.parse('schools')
ideation = df.parse('Ideation')
alldates = df.parse('All Dates')
```

```python
#schools_susptype = schools.merge(susptypefinal, left_on='DBN', right_on='ATS', how='left')
ideation_df = ideation.merge(alldates, left_on='Date', right_on='Date', how='left')
```

```python
#schools_susptype.drop('ATS', axis = 1, inplace = True)
#Drop Day of Week, School Day, Nearest School Day, 2 Days, 5 Days
ideation_df.drop('Day of Week', axis = 1, inplace = True)
ideation_df.drop('School Day', axis = 1, inplace = True)
ideation_df.drop('Nearest School Day', axis = 1, inplace = True)
ideation_df.drop('2 Days', axis = 1, inplace = True)
ideation_df.drop('5 Days', axis = 1, inplace = True)
ideation_df.drop('UpdatedText', axis = 1, inplace = True)
ideation_df.drop('EventType', axis = 1, inplace = True)
```

```python
ideation_df
```

```python
#Create fileds: Entered on Time: If OccDateEntered <= Final Day to Enter - 1 Day --> YES, otherwise, NO
#Source: https://www.geeksforgeeks.org/python-creating-a-pandas-dataframe-column-based-on-a-given-condition/
#df['hasimage'] = np.where(df['photos']!= '[]', True, False)  
#Source: https://www.dataquest.io/blog/tutorial-add-column-pandas-dataframe-based-on-if-else-condition/
#df['Price'] = [1500 if x =='Music' else 800 for x in df['Event']] 

ideation_df['Entered on Time'] = np.where(ideation_df['OccDateEntered'] <= ideation_df['Final Day to ENTER - 1 Day'], 'YES', 'NO')
```

```python
ideation_df
```

```python
#Updated on Time: If Updated Date <= 10 Days --> YES, otherwise, NO
#ideation_df['Entered on Time'] = np.where(ideation_df['OccDateEntered'] <= ideation_df['Final Day to ENTER - 1 Day'], 'YES', 'NO')
ideation_df['Updated on Time'] = np.where(ideation_df['Updated Date'] <= ideation_df['10 Days'], 'YES', 'NO')
```
